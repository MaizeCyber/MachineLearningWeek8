By far the most important feature was svcscan.nervices, followed by svcscan.process_services. This number represents the total number of Windows services found in a memory dump after running a service scan. This is considered to be a high-value indication of malware due to peristence mechanisms used in many types of malware. The creation or installations of new Windows Services for persistence often diverges from the baseline of many benign programs.
